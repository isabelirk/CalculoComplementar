**Problema 01**: Estima-se que a produção semanal de uma fábrica é dada pela função f(x,
y) = 1.200x+ 500y+ x
2 y – x
3

unidades, onde x é o número de operários
especializados e y o número de operários não especializados utilizados no trabalho. No
momento, a mão de obra disponível é constituída por 30 operários especializados e 60
operários não especializados. Use a análise marginal para estimar a variação da
produção se mais 1 operário especializado for contratado e o número de operários não
especializado permanecer constante.

```
#Analise Marginal
x_sym, y_sym, z_sym = sp.symbols('x y z')
f = 1200*x_sym + 500*y_sym + x**2*y_sym - x_sym**3 - y_sym**2

df = sp.diff(f, x_sym)
z = lambda x, y: eval(str(df))
print('Analise Marginal:', z(30,60))

#Taxa de Variação
fx = lambda x, y: 1200*x + 500*y + x**2*y - x**3 - y**2
print('Taxa de variação:', fx(30+1, 60)-fx(30,60))
```

empírica S(W, H) = 0,0072W0,425 H
0,725 , onde W (Kg) e H(cm), são, respectivamente, o
peso e a altura da pessoa. No momento, uma certa criança pesa 34Kg e tem 1m e 20cm

• Determine os dados do problema:


*   A criança pesa 34 kg.
*   A criança tem 1 m e 20 cm de altura.
*   O objetivo do problema é estimar qual será a variação da área superficial se a criança engorda 1 kg e altura permanecer inalterada.


Calcule as derivadas parciais Sw(50,163) e SH(50, 163) e interprete-as como taxas
de variação.

```
w_symbol, h_symbol = sp.symbols('w h')

w_sym, h_sym, z_sym = sp.symbols('w h z')
f = 0.0072*w_sym**0.425 + h_sym**0.725

#derivada parcial
df = sp.diff(f, w_sym, h_sym)
z = lambda w, h: eval(str(df))
print('Derivada parcial:', z(50,163))

#taxa de variacao
fw = lambda w, h: 1200*w + 500*h + w**2*h - w**3 - h**2
print('Taxa de variação:', fw(34+1, 120)-fw(34,120))
```
